```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import pandas as pd
data = pd.read_csv('/content/creditcard.csv')
print(data.columns)
if 'Class' in data.columns:
    x= data.drop('Class', axis=1) # Drop the column named 'Class'
    y = data['Class']
else:
    # Handle the case where neither 'target_variable' nor 'Class' exist
    print("Error: Neither 'target_variable' nor 'Class' column found in the DataFrame.")


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train, y_train)

# ipython-input-5-5efcbea844cf
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, f1_score
model = LogisticRegression()
# training
model.fit(x_smote,y_smote) # Now x_smote and y_smote are defined
# testing
y_pred = model.predict (x_test)
print(classification_report(y_test,y_pred))
print("F1 Score",f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, f1_score
model = RandomForestClassifier(n_jobs=-1)
# training
model.fit(x_smote,y_smote)
# testing
y_pred = model.predict (x_test)
print(classification_report(y_test,y_pred))
print("F1 Score",f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, f1_score
model = XGBClassifier(n_jobs=-1)
# training
model.fit(x_smote,y_smote)
# testing
y_pred = model.predict (x_test)
print(classification_report(y_test,y_pred))
print("F1 Score",f1_score(y_test, y_pred))
```
